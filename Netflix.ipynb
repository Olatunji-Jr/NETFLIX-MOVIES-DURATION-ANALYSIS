```python
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv("C:/Users/personal/Downloads/netflix_data.csv")
netflix_subset = netflix_df['show_id']
netflix_movies = netflix_df.loc[:,['title','country','genre','release_year','duration']]
short_movies = netflix_movies['duration'] < 60
#Use loop to add Colors
colors = []
for lab, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        colors.append('blue')
    elif row['genre'] == 'Documentaries':
        colors.append('green')
    elif row['genre'] == 'Stand-up' :
        colors.append('yellow')
    elif row['genre'] == 'Others' :
        colors.append('red')
    else :
        colors.append('pink')
fig = plt.scatter(netflix_df['release_year'],netflix_df['duration'], c=colors,alpha=0.2)
plt.ylabel('Duration (min)')
plt.xlabel('Release Year')
plt.title('Movie Duration by Year of Release')
plt.grid(True)
plt.show()

net_res=netflix_df.groupby('release_year')['duration'].mean()
net_res.plot(x='Year',y='Duration', kind='line',title='Movie Duration Determined',color='red')
plt.grid(True)
plt.show()
```
